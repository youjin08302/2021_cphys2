## Gradient Descent

gradient descent 방법을 적용하여 다음 함수의 최저점을 구하여라.$$ f(x) = \frac{1}{2}x^4 -3 x^3 $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,0.01)
def f(x):
    return 0.5*x**4 - 3*x**3
plt.ylim(-100,100)
plt.xlim(-5,7)
plt.plot(x,f(x))
```

```python
x = 5
x_old = 7
eps = 0.01
precision = 0.000001

def fp(x):
    return 2*(x**3)-9*(x**2)
```

```python
while (abs(x - x_old)>precision):
    x_old = x
    x = x_old - eps*fp(x_old)
    
x
```

## Diabetes (regression)

다음의 당뇨 데이터를 회귀 분석하여라. 나이, BMI, 혈압, Y (당뇨병 진행 정도) 데이터를 이용하여라.

1.sklearn을 이용하지 않고 (자신의 코드로)
2.sklearn을 이용해서

```python
from sklearn import datasets
```

```python
diabetes=datasets.load_diabetes()
```

```python
diabetes.keys()
```

```python
diabetes.DESCR
```

```python
X_age=diabetes.data[:,0] # 나이
X_bmi=diabetes.data[:,2] # BMI
X_p=diabetes.data[:,3] # 혈압
Y = diabetes.target
```

```python
plt.figure(figsize=(15,5))

plt.subplot(131)
plt.plot(X_age,Y,'x')
plt.subplot(132)
plt.plot(X_bmi,Y,'x')
plt.subplot(133)
plt.plot(X_p,Y,'x')
```

```python
def f1(a,b,Xi_age):
    return a*Xi_age+b

def mse1(a,b,Y):
    Y_hat1=(a*X_age)+b
    return np.mean((Y_hat1-Y)**2)
```

```python
eps=0.00001
n=0;
a=1;b=0

while(n<100000):
    if(mse1(a+eps,b,Y)>mse1(a-eps,b,Y)):
        a=a-eps
    else:
        a=a+eps
            
    if(mse1(a,b+eps,Y)>mse1(a,b-eps,Y)):
        b=b-eps
    else:
        b=b+eps
        
    n=n+1
```
